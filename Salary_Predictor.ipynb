```python
import pickle
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import load_model
```

```python
# Loading trained model and artifacts
model = load_model('SalaryPredictionModel.keras')

with open('OneHotEncoderGeography.pkl','rb') as file:
    OneHotEncoderGeography=pickle.load(file)

with open('LabelEncoderGender.pkl', 'rb') as file:
    LabelEncoderGender = pickle.load(file)

with open('StandardScaler.pkl', 'rb') as file:
    StandardScaler = pickle.load(file)
```

```python
# Example input data
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'Exited': 0
}

input_df=pd.DataFrame([input_data])
input_df
```

```python
input_df['Gender']=LabelEncoderGender.transform(input_df['Gender'])

geo_encoded = OneHotEncoderGeography.transform([[input_data['Geography']]])
geo_encoded_df = pd.DataFrame(geo_encoded, columns=OneHotEncoderGeography.get_feature_names_out(['Geography']))

input_df=pd.concat([input_df.drop("Geography",axis=1),geo_encoded_df],axis=1)
input_df

input_scaled=StandardScaler.transform(input_df)
input_scaled
```

```python
## Predict Salary
prediction=model.predict(input_scaled)
print(f'Predicted Salary: {prediction[0][0]}')
# prediction_proba = prediction[0][0]

# if prediction_proba > 0.5:
#     print('The customer is likely to churn.')
# else:
#     print('The customer is not likely to churn.')
```
